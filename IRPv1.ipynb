```python
import numpy as np
import pandas as pd
```

```python

file = open("/content/drive/MyDrive/Information Retrieval/IR.txt",'r')
data = file.read()
data = data[1:].split('.')
data = [i.strip() for i in data]
print(len(data))
print(data[3])
l = []
for i in data:
  if len(i)>0:
    splitted = i.split(' ')
    for j in splitted:
      l.append(j)

s = set(l)
t = list(s)
print(len(t))
df = pd.DataFrame(np.array([np.array([0]*len(s))]*(len(data)+2)),columns=s)
df


```

```python
for w in t:
  for d in range(len(data)):
    df.loc[d,w] = data[d].count(w)
    # print(df[t][d])
  docf= len(data)-list(df[w]).count(0)+2
  df.loc[len(data),w] = docf
  idf = np.log(len(data)/docf)
  df.loc[len(data)+1,w] = idf
df
```

```python
for i in range(len(data)):
  df.iloc[i] = df.iloc[i]*df.iloc[8]
  df.iloc[i]= df.iloc[i]/np.sqrt(sum(df.iloc[i]**2))
df
```

```python
q= "obtaining information system that is relevant"
q=q.split(' ')
dfq = pd.DataFrame(np.array([np.array([0]*len(t))]),columns=s)
dfq
for w in q:
  dfq.loc[0,w] = q.count(w)
dfq.iloc[0] = df.iloc[8]*dfq.iloc[0]
dfq.iloc[0]= dfq.iloc[0]/np.sqrt(sum(dfq.iloc[0]**2))
dfq
```

Cosine similarity

```python
result = []
for i in range(len(data)-1):
  s = np.dot(df.iloc[i],dfq.iloc[0])
  result.append(s)
print(result)
print(max(result))
print(data[result.index(max(result))])
```
